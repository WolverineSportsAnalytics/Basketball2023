```python
import csv
import json
tournament_name = '2006'
kenpom_csv_path = ''
kenpom_csv_file_name = '2006 Pomeroy College Basketball Ratings.csv'
json_path = ''
json_file_name = tournament_name + '.json'
```

```python
# function to convert a csv to json
def make_json_kenpom(csvFilePath, jsonFilePath):
    data = []
    # read csv reader 
    with open(csvFilePath, encoding='utf-8') as csvf:
        csvReader = csv.DictReader(csvf)
        for row in csvReader:
            content = {}
            name = row['Team']
            x = ''.join([i for i in name if not i.isdigit()])   
            content['Team'] = x.strip()
            content['Rank'] = row["Rk"]
            content['AdjEM'] = row['AdjEM']
            content['AdjO'] = row['AdjO']
            content['AdjD'] = row['AdjD']
            data.append(content)
            
    # open a json writer
    with open(jsonFilePath, 'w', encoding='utf-8') as jsonf:
        json_obj = json.dumps(data, indent = 4)
        print(json_obj)
        jsonf.write(json_obj)
```

```python
if __name__ == '__main__':
        make_json_kenpom(kenpom_csv_path + kenpom_csv_file_name, json_path + json_file_name)
    
```
